```python
# Importing the libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
# Reading the file
df = pd.read_csv('most_runs_in_cricket.csv')
```

```python
df.head()
```

```python
df= df.drop('Unnamed: 0',axis=1)
```

```python
df.head()
```

```python
df.dtypes
```

```python
def seperate_years(col):
    return col[0:4]

df['start_year'] = df['Span'].apply(seperate_years)
```

```python
def seperate_years2(col):
    return col[5:]

df['end_year'] = df['Span'].apply(seperate_years2)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['start_year'] = df['start_year'].astype('int32')
df['end_year'] = df['end_year'].astype('int32')

df['number_of_years_played'] = df['end_year'] - df['start_year']
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
def remove_plus_in_BF(col):
    if col[-1]=='+':
        return col[:-1]
    else:
        return col
df['BF'] = df['BF'].apply(remove_plus_in_BF)
```

```python
df.head()
```

```python
df['BF'] = df['BF'].astype('int')
```

```python
df.dtypes
```

```python
df.corr()
```

```python
plt.figure(figsize=(12,8),dpi=200)
sns.heatmap(df.corr(),annot=True)
```

## Observations

Therefore, in order to score more centuries all you have to do is play more innings, maintain good average and deal in 4s (6s have a very low correlation!).

2. Filtering the features further - Inn, Ave and 4s

```python
df = df[['Inns','Ave','4s','100']]
```

## Training the models

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 101)
```

## Multiple Linear Regression 

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_predict = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
n = len(X_test)
k = X_test.shape[1]
MAE = mean_absolute_error(y_test,y_predict)
MAPE = mean_absolute_percentage_error(y_test,y_predict)
MSE = mean_squared_error(y_test,y_predict)
RMSE = np.sqrt(MSE)
r2 = r2_score(y_test,y_predict)
adj_r2 = 1 - ((n-1)*(1-r2)/(n-k-1))
```

```python
print('MAE=',MAE,'\nMAPE=',MAPE,'\nMSE=',MSE,'\nRMSE=',RMSE,'\nr2=',r2,'\nadj_r2=',adj_r2)
```

```python
regressor.predict([[782,48.5,4076]])
```

## Support Vector Regression

```python
y_predict = svr_regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
n = len(X_test)
k = X_test.shape[1]
MAE = mean_absolute_error(y_test,y_predict)
MAPE = mean_absolute_percentage_error(y_test,y_predict)
MSE = mean_squared_error(y_test,y_predict)
RMSE = np.sqrt(MSE)
r2 = r2_score(y_test,y_predict)
adj_r2 = 1 - ((n-1)*(1-r2)/(n-k-1))
```

```python
print('MAE=',MAE,'\nMAPE=',MAPE,'\nMSE=',MSE,'\nRMSE=',RMSE,'\nr2=',r2,'\nadj_r2=',adj_r2)
```

```python
svr_regressor.predict([[782,48.5,4076]])
```
